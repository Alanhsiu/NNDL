## Spatial batch normalization

In fully connected networks, we performed batch normalization on the activations.  To do something equivalent on CNNs, we modify batch normalization slightly.  

Normally batch-normalization accepts inputs of shape `(N, D)` and produces outputs of shape `(N, D)`, where we normalize across the minibatch dimension `N`. For data coming from convolutional layers, batch normalization accepts inputs of shape `(N, C, H, W)` and produces outputs of shape `(N, C, H, W)` where the `N` dimension gives the minibatch size and the `(H, W)` dimensions give the spatial size of the feature map.

How do we calculate the spatial averages?  First, notice that for the `C` feature maps we have (i.e., the layer has `C` filters) that each of these ought to have its own batch norm statistics, since each feature map may be picking out very different features in the images.  However, within a feature map, we may assume that across all inputs and across all locations in the feature map, there ought to be relatively similar first and second order statistics.  Hence, one way to think of spatial batch-normalization is to reshape the `(N, C, H, W)` array as an `(N*H*W, C)` array and perform batch normalization on this array.

## Question

In our architecture, we apply batch normalization *before* ReLU. What would change if we applied it *after* ReLU instead? Which ordering do you think is better, and why?

## Answer

**What would change if we applied Batch Normalization after ReLU:**
If we apply Batch Normalization after ReLU, the input to the BN layer will be strictly non-negative (since ReLU outputs $\max(0, x)$). The distribution of these activations is non-Gaussian and heavily skewed, with a massive spike at zero. When BN attempts to zero-center this data, it will calculate a positive mean and subtract it from all values. As a result, the previously zeroed-out (deactivated) neurons will be shifted to negative values. This alters the fundamental sparse, non-negative property that ReLU was meant to introduce.

**Which ordering is better and why:**
Applying Batch Normalization **before** ReLU is generally considered the better approach (and is the standard practice recommended in the original Batch Normalization paper by Ioffe and Szegedy). 

Here is why:
1. **Appropriate Distribution for BN:** Batch Normalization assumes that the data it is normalizing is roughly symmetric and Gaussian. The output of an affine or convolutional layer (a weighted sum) fits this assumption much better than the truncated, skewed output of a ReLU layer. 
2. **Controlling Activation:** By applying BN before ReLU, the network ensures that the inputs to the activation function are stable and well-scaled. The learnable parameters ($\gamma$ and $\beta$) in the BN layer can then optimally shift and scale this symmetric distribution to control exactly what proportion of the values pass through the ReLU (i.e., are greater than zero).
3. **Preserving Sparsity:** ReLU introduces representational sparsity by completely turning off certain neurons (outputting exactly $0$). If BN is applied after ReLU, it destroys this sparsity by shifting those exact zeros into negative numbers, forcing the next layer to deal with a completely different input landscape.
